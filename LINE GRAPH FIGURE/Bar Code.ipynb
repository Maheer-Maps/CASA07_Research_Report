```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('bar_chart_data.csv')

# Rename the columns to simpler names
data.rename(columns={
    'Year 6': 'Year 6',
    'Average percentage of healthy children in Year 6 per borough': 'Year 6 Average',
    'Average percentage of healthy children in Reception per borough': 'Reception Average'
}, inplace=True)

# Filter the data to keep years 2019 and 2021 while excluding 2020
data_filtered = data[(data['Year 6'] != 2020)]

# Plot the data
plt.figure(figsize=(12, 8))

# Plot Year 6 Average with green color (continuous line)
plt.plot(data_filtered['Year 6'], data_filtered['Year 6 Average'], label='Year 6 Average', marker='o', color='green', linestyle='-')

# Plot Reception Average with default color (continuous line)
plt.plot(data_filtered['Year 6'], data_filtered['Reception Average'], label='Reception Average', marker='o')

# Customize the X-axis to label every year
plt.xticks(data_filtered['Year 6'])

# Add a "knook" at the bottom of the Y-axis to indicate it doesn't start at 0
plt.ylim(55, 80)
plt.gca().spines['bottom'].set_position(('outward', 10))
plt.gca().spines['left'].set_bounds(55, 80)

# Update axis labels and title
plt.title('Average percentage of healthy Year 6 and Reception children per year in London', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage of Healthy Children per London Borough', fontsize=12)

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
